```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics as st
from seaborn import FacetGrid




# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")
diamonds_data.head(5)
```

## STATISTICS challenge:

# calculate the mean of the price by cut

```python
mean_price_by_cut = diamonds_data.groupby('cut')['price'].mean()

print(mean_price_by_cut)
```

# calculate the standard deviation of the price by cut

```python
sd_price_by_cut = diamonds_data.groupby('cut')['price'].std()

print(sd_price_by_cut)
```

# calculate the median of the price by cut

```python
median_price_by_cut = diamonds_data.groupby('cut')['price'].median()

print(median_price_by_cut)
```

```python
# calculate the mode of the price by cut
mode_price_by_cut = diamonds_data.groupby('cut')['price'].apply(lambda x: x.mode().iloc[0])

print(mode_price_by_cut)
```

```python
# calculate the variance of the price by cut
variance_price_by_cut = diamonds_data.groupby('cut')['price'].var()

print(variance_price_by_cut)
```

```python
# Calculate the 25th and 75th percentiles of the price by cut
percentile_25 = diamonds_data.groupby('cut')['price'].quantile(0.25)
percentile_75 = diamonds_data.groupby('cut')['price'].quantile(0.75)

print("25th percentile:")
print(percentile_25)
print("\n75th percentile:")
print(percentile_75)
```

```python
# Calculate the range of the price by cut
price_range_by_cut = diamonds_data.groupby('cut')['price'].apply(lambda x: x.max() - x.min())

print(price_range_by_cut)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=diamonds_data)
plt.title('Boxplot of Diamond Price Vs Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```

```python
#Barplot: Average Diamond Price by Color
plt.figure(figsize=(10, 6))
sns.barplot(x='color', y='price', data=diamonds_data, estimator=np.mean)
plt.title('Average Diamond Price by Color')
plt.xlabel('Color')
plt.ylabel('Average Price')
plt.show()
```

```python
diamonds=sns.load_dataset('diamonds')
g=FacetGrid(diamonds,col="cut", hue="cut", col_wrap=3)
g.map(plt.scatter, "carat", "price",alpha=0.7)
g.set_titles("{col_name}")
g.set_axis_labels("Carat","Price")
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Scatterplot with Facets: Scatterplot of Carat vs. Price by Cut")
plt.show()
```

```python
cut_counts = diamonds_data['cut'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(cut_counts, labels=cut_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Diamond Cut')
plt.show()
```
